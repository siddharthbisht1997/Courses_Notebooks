# Experiment 1
##### Fundamentals of Quantitative Modeling

*Aim :  To see the change in the Target Variable w.r.t. the Predictor Variable for the following functions.*
1.   Linear $y = mx + c$
2.   Power $y = x^m$
3.   Exponential $y = e^{mx}$
4.   Logarithm $y = log_ex$

```
import numpy as np
import matplotlib.pyplot as plt
```

```
linear = lambda x,m,c : m*x + c
power = lambda x,m : np.power(x,m)
exp = lambda x,m : np.exp(m*x)
log = lambda x : np.log(x)
```

```
def plot_func(ax,i,j,x,y,title):
  ax[i][j].plot(x,y)
  ax[i][j].set_xlabel("x")
  ax[i][j].set_ylabel("y")
  ax[i][j].set_title(title)
```

```
x = np.arange(0.1,5.1,0.1)

fig, ax = plt.subplots(2,3)
fig.set_size_inches(20,10)

plot_func(ax,0,0,x,linear(x,1,1),"Linear m=1,c=1")
plot_func(ax,0,1,x,power(x,2),"Power m=2")
plot_func(ax,0,2,x,exp(x,1),"Exponent m=1")
plot_func(ax,1,0,x,log(x),"Natural Log")
plot_func(ax,1,1,x,exp(x,-1),"Exponent m=-1")
plot_func(ax,1,2,x,power(x,-1),"Power m=-1")
plt.show()
```

```
def plot_change(ax,i,j,x,title):
  ax[i][j].plot(x)
  ax[i][j].set_xlabel("#")
  ax[i][j].set_ylabel("change")
  ax[i][j].set_title(title)
```

```
x1 = np.arange(0.1,5.1,0.1)
x2 = x1 + 1

fig, ax = plt.subplots(2,3)
fig.set_size_inches(20,10)

plot_change(ax,0,0,linear(x2,2,3) - linear(x1,2,3),"Linear m=2,c=3")
plot_change(ax,0,1,power(x2,-1)-power(x1,-1),"Power m=-1")
plot_change(ax,0,2,power(x2,2)-power(x1,2),"Power m=2")
plot_change(ax,1,0,exp(x2,1)-exp(x1,1),"Exponent m=1")
plot_change(ax,1,1,exp(x2,-1)-exp(x1,-1),"Exponent m=-1")
plot_change(ax,1,2,log(x2)-log(x1),"Natural Log")
plt.show()
```

# Experiment 2
##### Fundamentals of Quantitative Modeling

*Aim :  Linear(Additive Change) vs Proportionate Change in SI and CI*

If $P$ is the Principal Amount, $R$  is Rate of Interest, $T$ is the Time (in years) then

1. For $SI$,  $Value = P + PRT$ --> Linear in Time
2. For $CI$, $Value = P(1+R/100)^T$ --> Exponential in Time


```
import numpy as np
import matplotlib.pyplot as plt
```

```
V_si = lambda p,r,t : p + p*r*t
V_ci = lambda p,r,t : p*np.power(1 + r,t)
```

```
p = 100
r = 0.1
time = np.arange(0,30,1)
```

```
amount_si, amount_ci = [],[]

for t in time:
  amount_si.append(V_si(p,r,t))
  amount_ci.append(V_ci(p,r,t))
```

```
plt.figure(figsize=(10,6))
plt.plot(time,amount_si)
plt.plot(time,amount_ci)
plt.legend(['SI','CI'])
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Additive vs Proportionate Change")
plt.show()
```
